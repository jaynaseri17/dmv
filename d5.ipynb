```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("AirQuality.csv")
```

```python
print(data.head())
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
data.dropna(inplace=True)
cleaned_data = data.dropna()
```

```python
print(cleaned_data)
```

```python
print(cleaned_data.isnull().sum())
```

```python
print(cleaned_data['Date'].shape)
print(cleaned_data['AQI'].shape)
print(cleaned_data.dtypes)
```

```python
cleaned_data['Date'] = pd.to_datetime(cleaned_data['Date'])
```

```python
cleaned_data['AQI'] = pd.to_numeric(cleaned_data['AQI'], errors='coerce')
```

```python
plt.figure(figsize=(12, 6))
plt.xlabel('Date')
plt.ylabel('AQI')
plt.title('AQI Trend Over Time')
plt.plot(cleaned_data['Date'].to_numpy(), cleaned_data['AQI'].to_numpy(), label='AQI', color='blue', linewidth=2)
plt.legend()
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
print(data['Date'].shape)
print(data['PM2.5'].shape)
print(data['PM10'].shape)
print(data['CO'].shape)
print(data.dtypes)
```

```python
data['PM2.5'] = pd.to_numeric(data['PM2.5'], errors='coerce')
data['PM10'] = pd.to_numeric(data['PM10'], errors='coerce')
data['CO'] = pd.to_numeric(data['CO'], errors='coerce')
```

```python
plt.figure(figsize=(12, 6))
plt.title('Pollutant Levels Over Time')
plt.xlabel('Date')
plt.ylabel('Pollutant Concentration')
plt.plot(data['Date'].to_numpy(), data['PM2.5'].to_numpy(), label='PM2.5', color='red')
plt.plot(data['Date'].to_numpy(), data['PM10'].to_numpy(), label='PM10', color='green')
plt.plot(data['Date'].to_numpy(), data['CO'].to_numpy(), label='CO', color='purple')
plt.legend()
plt.show()
```

```python
cleaned_data['Date'] = pd.to_datetime(cleaned_data['Date'])

# Group the data by 'Date' and calculate the mean AQI for each date
aqi_by_date = cleaned_data.groupby('Date')['AQI'].mean().reset_index()

# Create a bar plot to compare AQI values across dates
plt.figure(figsize=(12, 6))
plt.bar(aqi_by_date['Date'], aqi_by_date['AQI'], color='blue', alpha=0.7)
plt.xlabel('Date')
plt.ylabel('Average AQI')
plt.title('Average AQI Across Dates')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pollutants = ['PM2.5', 'PM10', 'CO']


# Create violin plots for the same analysis (if desired)
plt.figure(figsize=(12, 6))
sns.violinplot(data=cleaned_data, x='AQI', y='PM2.5', palette='Set2', color='red')
plt.xlabel('AQI')
plt.ylabel('PM2.5')
plt.title('Violin Plot: Distribution of AQI for PM2.5')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(data=cleaned_data, x='AQI', y='PM10', palette='Set2', color='red')
plt.xlabel('AQI')
plt.ylabel('PM10')
plt.title('Violin Plot: Distribution of AQI for PM10')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(data=cleaned_data, x='AQI', y='CO', palette='Set2', color='red')
plt.xlabel('AQI')
plt.ylabel('CO')
plt.title('Violin Plot: Distribution of AQI for CO')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


# Define variables for the bubble chart
x = cleaned_data['PM2.5']
y = cleaned_data['CO']
bubble_size = cleaned_data['PM10']  # Represent 'PM10' using bubble size
aqi_values = cleaned_data['AQI']

# Create the bubble chart
plt.figure(figsize=(12, 8))
plt.scatter(x, y, s=bubble_size, c=aqi_values, cmap='coolwarm', alpha=0.7)
plt.xlabel('PM2.5')
plt.ylabel('CO')
plt.title('Bubble Chart: AQI vs. PM2.5 and CO')
plt.colorbar(label='AQI')  # Add colorbar to show AQI values
plt.grid(True)
plt.tight_layout()
plt.show()
```
